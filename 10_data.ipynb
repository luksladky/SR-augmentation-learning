```python
%nbdev_export_internal
from pathlib import Path
from fastai import *
from fastai.vision import *
from srthesis.data_div2k import *
from srthesis.data_realsr import *
```

```python
%nbdev_export
_all_ = ['test_patches']
```

```python
%nbdev_export
def get_data(y_size=128, bs=8, 
             dataset='realsr',
             dihedral_augs=True,
             repeat_train = 1, 
             max_train_len:int=sys.maxsize) -> ImageDataBunch:
    if dataset == 'realsr':
        return get_data_realsr(y_size, bs, dihedral_augs=dihedral_augs, 
                               repeat_train=repeat_train, 
                               max_train_len=max_train_len)
    if dataset == 'realsr_inverse':
        return get_data_realsr(y_size, bs, dihedral_augs=dihedral_augs, 
                               input_type = 'HR', output_type='LR2', 
                               repeat_train=repeat_train, 
                               max_train_len=max_train_len)
    if dataset == 'div2k':
        return get_data_div2k(y_size, bs)
```
